### Label-based Indexing and slicing with loc

### loc
`primarly loc[] is used for label based filtering([row, column])`<br>
`It can also be used using boolean filtering`

> Filtered boolean should be in the form of array/series only

```python
# summer data head

dd.head()
```

```python
# Select index[0] ('HAJOS, Alfred') by name/label and column[0] ('Year') by name only

dd.loc['HAJOS, Alfred','Year']  # .loc is used to select rows and columns using labels
```

```python
# Select index[0] ('HAJOS, Alfred') by name/label with all columns

dd.loc['HAJOS, Alfred']
```

```python
# Select index by name/label with all columns

dd.loc[['HAJOS, Alfred','MALOKINIS, Ioannis','CHASAPIS, Spiridon']]
```

```python
# Select index by name/label

dd.loc[['HAJOS, Alfred','MALOKINIS, Ioannis','CHASAPIS, Spiridon'],['City','Sport']]
```

```python
# Select index by name/label

dd.loc[['CHASAPIS, Spiridon'],'Sport':'Gender']
```

```python
# Select index by name/label
# Error due to duplicate index

dd.loc['HAJOS, Alfred':'CHASAPIS, Spiridon','Sport':'Gender']
```

```python
# filter data

dd.index.str.contains('HAJOS')
```

```python
# view filter data

dd.loc[dd.index.str.contains('HAJOS')]
```
