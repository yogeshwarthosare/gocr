```python
# IMport Libraries

import pandas as pd
import numpy as np
%matplotlib notebook
# %matplotlib widget
%matplotlib ipympl

    
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
from sklearn.linear_model import LinearRegression
```

```python
X, y = make_regression(n_samples= 100, n_features= 1, n_informative= 1, n_targets= 1, noise= 30, random_state= 0)
X
```

```python
# instance

lr = LinearRegression()
lr
```

```python
# fit model

lr.fit(X, y)
```

```python
lr.coef_, lr.intercept_
```

```python
#

b = 100
m = 43.087
lrate = 0.005

intercept = []
loss = []
epochs = 30 # No of iteration

for iter in range(epochs):
    slope = 0
    error = 0
    for i in range(X.shape[0]):
        slope = slope - 2 * (y[i] - (m * X[i]) - b)
        error = error + (y[i] - (m * X[i]) - b)**2
    b = b - (lrate * slope)
    intercept.append(b)
    loss.append(error)
    y_pred = m*X+b
    plt.plot(X, y_pred)
plt.scatter(X, y)
```

```python
plt.scatter(X, y)
```

```python
intercept = np.array(intercept).ravel()
```

```python
loss = np.array(loss).ravel()
loss
```

```python
fig, ax = plt.subplots(figsize=(9,5))
x_i = np.arange(-3,3,0.1)
y_i = x_i*m - 150
ax.scatter(X,y)
line, = ax.plot(x_i,x_i*50 - 4, 'r-', linewidth = 2)

def update(i):
    label = {'epoch {0}'.format(i+1)}
    line.set_ydata(x_i*m + intercept[i])
    ax.set_xlabel(label)

anim = FuncAnimation(fig, update, repeat = True, frames = epochs, interval = 1000)
```
